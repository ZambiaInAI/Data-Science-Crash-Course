# 2. Data Exploration Using Python

## 2.1. Getting the Data

```python
import os
import pandas as pd


HOUSING_PATH=os.path.join("datasets","housing")
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path=os.path.join(housing_path,"housing.csv")
    return pd.read_csv(csv_path)
```

## 2.2. Understanding the structure of Data

```python
housing=load_housing_data()
housing.head()
```

```python
#Getting a quick descritpion of our data. in particular the total number of rows and each attributes
#types and nummber of non-null values
housing.info()
```

```python
#We want to find the categories that belong to ocean_proximity
housing["ocean_proximity"].value_counts()
```

```python
#Getting the descriptive stats about our data
#Percentile indicates the value below which a given percentage of 
#observation of observation in a group of observations falls
housing.describe()
```

```python
from matplotlib import pyplot as plt

housing.hist(bins=50, figsize=(20,15))
plt.show()
```

__Sampling:__ is the selection of a subset from within a poplation to estimate the characteristics of a population.
- _Stratified sampling_ the population is divided into homogenous subgroups called strata and the right instances from is samples from each stratum to guarantee the  test set is a representative of the overall population

```python
import numpy as np
#Data scaling
#creating the median income category. 
#NB: medan income in numeric
housing["income_cat"]=np.ceil(housing["median_income"]/1.5)
#grouping everythin greater than 5
housing["income_cat"].where(housing["income_cat"]<5,5.0,inplace=True)
#Tail depicts outliers
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split=StratifiedShuffleSplit(n_splits=1, test_size=0.2,random_state=42)
for train_index,test_index in split.split(housing, housing["income_cat"]):
    strat_train_set=housing.loc[train_index]
    strat_test_set=housing.loc[test_index]
```

```python
housing.head()
```

```python
split=StratifiedShuffleSplit(n_splits=1, test_size=0.2,random_state=42)
for train_index,test_index in split.split(housing, housing["income_cat"]):
    strat_train_set=housing.loc[train_index]
    strat_test_set=housing.loc[test_index]
```

```python
strat_test_set["income_cat"].value_counts()/len(strat_test_set)
```

```python
for set_ in (strat_train_set,strat_test_set):
    set_.drop("income_cat",axis=1,inplace=True)
```
